```python
import pandas as pd,numpy as np,os,matplotlib.pyplot as plt,seaborn as sns
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv(r'E:\Batch 52- Day12\Position Salary .csv')
```

```python
x=df.iloc[:,[1]].values
y=df.iloc[:,[2]].values
```

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
rg.fit(x,y)
```

```python
rg.predict(x)
```

```python
r2_score(y,rg.predict(x))
```

```python
plt.plot(x,y)
plt.plot(x,rg.predict(x))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x,y)
```

```python
dt.predict(x)
```

```python
r2_score(y,dt.predict(x))*100
```

```python
plt.plot(x,y)
plt.scatter(x,dt.predict(x))
```

```python
x_grid = np.arange(min(x),max(x),0.01)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color= 'r')
plt.plot(x_grid, dt.predict(x_grid))
```

```python
plt.plot(x_grid,dt.predict(x_grid))
```

```python
import pandas as pd,numpy as np,os,matplotlib.pyplot as plt,seaborn as sns
from sklearn.metrics import r2_score
```

```python
dtf=pd.read_excel(r'E:\Batch 52- Day11\CPU Daily.xlsx')
dtf
```

```python
x=dtf.iloc[:,[1]].values
y=dtf.iloc[:,2].values
```

```python
from sklearn.ensemble import RandomForestRegressor
rfg=RandomForestRegressor(n_estimators=50)
rfg.fit(x,y)
```

```python
r2_score(y,rfg.predict(x))
```

```python
plt.plot(y)
plt.plot(rfg.predict(x))
```

```python
x_grid = np.arange(min(x),max(x),0.01)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color= 'r')
plt.plot(x_grid, rfg.predict(x_grid))
```

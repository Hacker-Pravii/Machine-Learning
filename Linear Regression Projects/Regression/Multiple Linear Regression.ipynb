```python
import numpy as np,pandas as pd,os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
df=pd.read_csv(r'50_Startups.csv')
df
```

```python
x=df.iloc[:,:4].values
y=df.iloc[:,-1:].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('EN',OneHotEncoder(),[3])],remainder='passthrough')
print(ct)
nx=ct.fit_transform(x)
fx=nx
```

```python
from sklearn.preprocessing import LabelEncoder as le
x[:,3]=le().fit_transform(x[:,3])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression as lrm
rg=lrm().fit(x_train,y_train)
r2_score(y_test,rg.predict(x_test))
```

```python
a=pd.DataFrame(nx,columns=['California','Florina','New York','R&D Spend','Administration','Marketing Spend'])
b=pd.DataFrame(y,columns=['Profit'])
c=pd.DataFrame(rg.predict(nx),columns=['Predicted Profit'])
d=pd.DataFrame(y-rg.predict(nx),columns=['Difference Profit vs Predicted Profit'])
r=pd.concat([a,b,c,d],axis=1)
r.to_csv('Predicted_salary',index=False)
```

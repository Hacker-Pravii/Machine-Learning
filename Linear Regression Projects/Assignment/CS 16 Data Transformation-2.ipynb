```python
import numpy as np,pandas as pd,os
import matplotlib.pyplot as plt
import seaborn as sns
x=pd.read_csv('DSScoreTerm1.csv')
y=pd.read_csv('MathScoreTerm1.csv')
z=pd.read_csv('PhysicsScoreTerm1.csv')
x=x.drop(['Ethinicity','Name'],axis=1)
y=y.drop(['Ethinicity','Name'],axis=1)
z=z.drop(['Ethinicity','Name'],axis=1)
z[z.Score==np.nan]=0
x[x.Score==np.nan]=0
y[y.Score==np.nan]=0
x.Score = x.Score.replace(np.nan, 0)
y.Score = y.Score.replace(np.nan, 0)
z.Score = z.Score.replace(np.nan, 0)
x
```

```python
a=pd.concat([x,y,z])
n=np.array(a)
for i in range(len(n[:,3])):
    if n[i][3]=="M":
        n[i][3]=1
    else:
        n[i][3]=2
n=pd.DataFrame(n)
n = n.set_axis([i for i in x.columns], axis=1)
n = n.rename_axis(None, axis=0)
n.to_csv('ScoreFinal1.csv',index=False)
n
```
